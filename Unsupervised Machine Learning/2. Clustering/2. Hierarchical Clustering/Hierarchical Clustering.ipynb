Hierarchical clustering, also known as hierarchical cluster analysis, is an algorithm that groups similar objects into groups called clusters. The endpoint is a set of clusters, where each cluster is distinct from each other cluster, and the objects within each cluster are broadly similar to each other.

#### <font color = blue>Import the Libraries

```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
warnings.filterwarnings("ignore")
```

#### <font color = blue>Import the Dataset

```python
dataset=pd.read_csv('Dataset/Mall_Customers.csv')  
dataset.head()
```

#### <font color = blue> Predictor Attribute

```python
X = dataset.iloc[:, [3,4]].values   # we have only X here becuase we are working Unsupervised data
```

![Machine Learning Project](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

## <font color = #950CDF> Part 2: </font> <font color = #4854E8> Build the Hierarchical Clustering</font>
In this part, we build the Hierarchical Clustering using Scikit-Learn

![ML](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

#### <font color = blue> Import Hierarchical from sklean

```python
import scipy.cluster.hierarchy as sch
```

#### <font color = blue> let's plot dendrogram 
dendrogram method is basically helps us to find optimal number of clusters

```python
# import the scipy package for dendogram
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram( sch.linkage(X, method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Customer')
plt.ylabel('Euclidean Distance')
plt.show()
```

#### <font color = blue> Build Hierachical with optimal no. of Clusters

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering( n_clusters= 5 , affinity= 'euclidean', linkage= 'ward')
hc
```

![Machine Learning Project](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

## <font color = #950CDF> Part 3: </font> <font color = #4854E8> Predict and Visualize the Result</font>

![ML](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

#### <font color = blue> Making a Prediction

```python
y_hc = hc.fit_predict(X)
```

#### <font color = blue> Visulising the Result

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c='red', label = 'Careful')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c='blue', label = 'Standard')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c='green', label = 'Target')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s=100, c='cyan', label = 'Careless')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s=100, c='magenta', label = 'Sensible')
plt.title('Cluster of Client')
plt.xlabel('Annual income (k$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
